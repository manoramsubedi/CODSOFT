```python
sns.JointGrid(data=iris_data,  x='petal_width',y='petal_length', hue="species").plot_joint(sns.scatterplot).plot_marginals(sns.boxplot)
plt.xlabel('petal width')
plt.ylabel('petal length')

sns.JointGrid(data=iris_data,  x='sepal_width',y='sepal_length', hue="species").plot_joint(sns.scatterplot).plot_marginals(sns.boxplot)
plt.xlabel('sepal width')
plt.ylabel('sepal length')
```

```python
x=iris_data.iloc[:,:-1]
y= iris_data.iloc[:,-1]
x, y
```

```python
#splitting data
from sklearn.model_selection import train_test_split
X_train ,X_test,y_train,y_test= train_test_split(x,y,test_size=0.2 , random_state=30)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#import logistic regression
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
pred = lm.predict(X_test)
```
